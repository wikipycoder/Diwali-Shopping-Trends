```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status", "unnamed1"], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.rename(columns={"Marital_Status": "Shadi"})
```

```python
df["Amount"] = df["Amount"].astype(int)
```

```python
df.describe()
```

```python
df["Amount"].describe()
```

```python
df.columns
```

```python
ax = sns.countplot(x="Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_palette = {"M": "blue", "F": "green"}
```

```python
sales_gender = df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
ax = sns.barplot(x="Gender", y="Amount", data=sales_gender, hue="Gender")
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Show the plot
plt.show()
```

```python
ax = sns.countplot(x="Age Group", hue="Gender", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False) 
sns.barplot(x="Age Group", y="Amount", data=sales_age, hue="Amount")
```

From the above we can the see that the most buyers are in the age between 26-35

```python
sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={"figure.figsize": (17, 5)})
sns.barplot(x="State", y="Orders", data=sales_state)
```

From the above graph we can tell that the most of the orders placed are from Uttar Pardesh state

```python
df.columns
```

```python
sns.set(rc={"figure.figsize": (7, 5)})
ax = sns.countplot(x="Marital_Status", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
marital_sales = df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
ax = sns.barplot(x="Marital_Status", y="Amount", hue="Gender", data=marital_sales)
for bars in ax.containers:
    ax.bar_label(bars)
```

from the above graph we can see the most of the buyers are married women and they have high purchasing power

```python
sns.set(rc={"figure.figsize": (17, 4)})
ax = sns.countplot(x="Occupation", data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_sales = df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x="Occupation", y="Amount",  data=occupation_sales)
```

*From above graph we can comprehend that most of the people who have high purchasing power are in the IT sector*


```python
sns.set(rc={"figure.figsize": (23, 5)})
sns.countplot(x="Product_Category", data=df)
```

```python
sales_product_category = df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={"figure.figsize": (23, 4)})
sns.barplot(x="Product_Category", y="Amount", data=sales_product_category)
```

From the above graph we can understand that most sold products are from food, clothing and electronics category

```python
## **Conclusion**

Married women of 26-35 years age group from UP, Mahrashtra, and Karnatakka working in IT, Healthcare, and Aviation are more likely to buy products
Food Clothing and Electronics Category.
```
